```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
#checks whether the file exists

os.path.exists('airline_accidents.csv.zip')
```

```python
df=pd.read_csv('airline_accidents.csv.zip',low_memory=False)
```

```python
df.head()
```

```python
df.columns
```

```python
#removes the  unimportant information not required for calculations

drop_col=['Event Id','Airport Code','Airport Name','Registration Number','FAR Description','Schedule','Purpose of Flight']
df.drop(columns=drop_col,inplace=True)
```

```python
df.head()
```

```python
#combines the total number of injuries in a column

injury=['Total Fatal Injuries','Total Serious Injuries', 'Total Minor Injuries']
df['all injured']=df[injury].sum(axis=1)
```
